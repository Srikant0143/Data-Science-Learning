```python
# <b>merge</b>
# inner ---> default
# left
# right
# outer
```

```python
df1=pd.DataFrame({"Name":["kumar","dan","kennady","Ram"],"Age":[23,67,89,62]})
df1
```

```python
df2=pd.DataFrame({"Name":["lena","dan","kennady","Ram"],"Age":[23,67,89,56]})
df2
```

```python
pd.merge(df1,df2)   # by default it will take inner join
```

```python
pd.merge(df1,df2, on="Age")   # inner join by default
```

```python
pd.merge(df1,df2, on="Name")   # inner join by default
```

```python
pd.merge(df1,df2, on="Name", how="left")  # left join is data of df1 & df2 from left side which is df1
```

```python
pd.merge(df1,df2, on="Name", how="right")   # left join is data of df1 & df2 from right side which is df2
```

```python
pd.merge(df1,df2, on="Name", how="outer")
```
