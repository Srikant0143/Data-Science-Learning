```python
import numpy as np
from scipy import stats
```

```python
salary = np.array([10,25,67,35,89,90])
age = np.array([15,25,23,45,67,42])
```

```python
coeff,pval = stats.pearsonr(salary,age)
print("coefficient", coeff)
print("Pvalue is", pval)
```

```python
if pval < 0.05:
    print("Alternate hypo--- Relation exists")
else:
    print("Null hypo-- no relation exists")
```

```python
coeff,pval = stats.spearmanr(salary,age)
print("coefficient", coeff)
print("Pvalue is", pval)
```

```python
import pandas as pd
import seaborn as sns
```

```python
dataset = sns.load_dataset("tips")
dataset
```

```python
data = pd.crosstab(dataset["sex"],dataset["smoker"])
data
```

```python
help(stats.chi2_contingency)
```

```python
chi,pval,dof,exp=stats.chi2_contingency(data.values) # data.values
print("Chisquare", chi)
print("plvaue", pval)
print("Expected", exp)
```

```python
if pval < 0.05:
    print("Alternate hypo--- Relation exists")
else:
    print("Null hypo-- no relation exists")
```
