# Data Visualization using seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
data1=sns.load_dataset("tips")
data1
```

```python
data1.info()
```

```python
# scatter plot
plt.figure(figsize=(10,5))
sns.scatterplot(x="total_bill", y="tip", data=data1, hue="sex")
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x="total_bill", y="tip", data=data1, hue="day")
plt.savefig("demo.png")
```

```python
# barlpot
sns.barplot(x=data1["sex"], y=data1["total_bill"])
```

```python
sns.barplot(x=data1["sex"], y=data1["total_bill"], estimator=np.max)
```

```python
sns.barplot(x=data1["sex"], y=data1["total_bill"], palette="husl")
```

```python
sns.barplot(x=data1["sex"], y=data1["total_bill"], palette="husl",
            hue=data1["day"], ci=None)
```

```python
data1["total_bill"].min()
```

```python
data1["total_bill"].max()
```

```python
# count plot   ---uni-variable
data1["day"].value_counts()
```

```python
sns.countplot(x=data1["day"], hue=data1["sex"])
```

```python
data1.info()
```

```python
sns.countplot(x=data1["day"], hue=data1["time"])
```

```python
# strip plot
# categorical and numerical
sns.stripplot(x=data1["sex"],y=data1["total_bill"], hue=data1["day"])
```

```python
sns.swarmplot(x=data1["sex"],y=data1["total_bill"], hue=data1["day"])
```

```python
# pair plot

sns.pairplot(data1, hue="sex")
plt.savefig("sns.png")
```
