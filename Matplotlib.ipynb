```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
t=np.linspace(0,20,10)
t
```

```python
y=t**2
y
```

```python
plt.plot(t,y,c="b", label="Wave")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Waveform", size=20)
plt.legend()
```

```python
plt.plot(t,y,c="b",alpha=1, ls="-.", marker="+",markersize="10")
```

```python
# multiple functions in same window
plt.plot(t,np.sin(t), color="g", label="sine")
plt.plot(t,np.cos(t), color="b", label="cos")
plt.legend()
```

```python
plt.plot(t,np.sin(t), color="g", label="sine")
plt.legend()
plt.show()
plt.plot(t,np.cos(t), color="b", label="cos")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(4,12))
plt.subplot(4,1,1)
plt.plot(t,np.sin(t), c="r", label="Sine")
plt.legend()

plt.subplot(4,1,2)
plt.plot(t,np.cos(t), c="g", label="Cos")
plt.legend()

plt.subplot(4,1,3)
plt.plot(t,np.tan(t), c="k", label="tan")
plt.legend()

plt.subplot(4,1,4)
plt.plot(t,np.sinh(t), c="b", label="Sinh")
plt.legend()
```

```python
plt.scatter(t,np.sin(t), color="g", label="sine")
```

```python
plt.bar(t,np.sin(t), color="g", label="sine")
```

You have been provided with a dataset that lists Ohio State’s leading causes of death from the year 2012.

Using the two data points:

Cause of deaths and
Percentile
Draw a pie chart to visualize the dataset

```python
Cause= ["Chronic Disease", "Unintetional Injuries", "Alzheimers", "Infuenza and Pneumonia", "Sepsis", "Others"]
Percentile = [62,5,4,2,1,26]
```

```python
plt.figure(figsize=(7,5))
plt.pie(Percentile, labels=Cause, autopct="%.0f%%", explode=[.1,0,0,0,0,0])
plt.title("cause of deaths", size=25)
plt.show()
```
